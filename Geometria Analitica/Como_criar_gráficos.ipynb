# Criação dos Vetores

Para a utilização das equações para a criação dos gráficos, temos que fazer com que os eixos utilizados sejam vetores, isso significa que se temos uma equação como y = x^2, temos que:

x = [0, 1, 2, ..., 8, 9, 10]

y = [0, 1, 4, ..., 64, 81, 100]

Porque para essa equação, todo valor de x possui um valor y que está elevado ao quadrado. 

Ok. Mas não queremos ter que fazer valor por valor em todas as equações. Temos uma maneira mais fácil de fazer isso. 

Primeiramente, importamos as bibliotecas das quais falamos antes. Deixamos em um bloco separado para importar todas de uma vez e facilitar a leitura. Rodamos esse código utilizando Ctrl + Enter (com o bloco selecionado) ou clicando em cima do botão de play do bloco.






````
import numpy as np
import matplotlib.pyplot as plt
````

A criação dos vetores é feita dessa forma. O eixo x se torna um vetor que
varia de -100 até 100. Enquanto o eixo y recebe a equação em função de x 
que iremos utilizar.

````
x = np.array(range(-100, 100))
y = x ** 2

# Cria o gráfico e adiciona as váriaveis a serem utilizadas
plt.plot(x,y)

# Adiciona um título ao gráfico (opcional)
plt.title('Gráfico de y = x²')

# Adiciona nomes para os eixos (opcional)
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')

'''
Caso queira deixar o gráfico em um estilo parecido com o plano cartesiano.
Infelizmente é bem trabalhoso fazer um plano cartesiano que fique identico ao
que fazemos no papel ou no geogebra, mas dá pra ter uma noção com esse. 
(opcional)
'''
plt.grid(alpha=.4,linestyle='--')

''' 
Mostra o gráfico na tela. Na verdade, essa linha não é muito necessária, porém
sem ela é gráfico vem com uma linha de código com o tipo de gráfico e da biblioteca
que estamos usando. Para deixar tudo mais limpo, utilizamos essa linha abaixo.
'''
plt.show()
````

Com isso é possível criar uma infinidade de gráficos. 

Note os códigos e a diferença entre eles. Veja o que cada coisa faz e o que você pode fazer com essa ferramenta.

````
x = np.array(range(-100, 100))
y = x ** 3
plt.plot(x,y)
plt.title('Gráfico de y = x³')
plt.grid(alpha=.4,linestyle='--')
plt.show()
````

````
x = np.array(range(-100, 100))
y = x * np.sin(x)
plt.plot(x,y)
plt.title('Gráfico de y = x * sen(x)')
plt.grid(alpha=.4,linestyle='--')
plt.show()
````

````
x = np.array(range(-100, 100))
y = x * np.cos(x)**2
plt.plot(x,y)
plt.title('Gráfico de y = x * cos(x)²')
plt.grid(alpha=.4,linestyle='--')
plt.show()
````

````
x = np.array(range(-100, 100))
y = x * np.exp(x)
plt.plot(x,y)
plt.title('Gráfico de y = x * e^x')
plt.grid(alpha=.4,linestyle='--')
plt.show()
````

````
x = np.array(range(-100, 100))
y = x / (1 + x**2)
plt.plot(x,y)
plt.title('Gráfico de y = x / (1 + x²)')
plt.grid(alpha=.4,linestyle='--')
plt.show()
````
